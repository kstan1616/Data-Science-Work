*** Exercise 1: ***
Plot the following functions in a 4x4 block using lambda expressions, pyplot.subplots and figure.subplots_adjust (only use the matplotlib libary). Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title. Which one of the functions is a one-to-one map?
1. $F_1(x) = Exp(-0.5*(x-2)^2)$
2. $F_2(x) = x^2 * Exp(-0.8*(x+2)^2)$
3. $F_3(x) = F_2(x)*F_1(x)$
4. $F_4(x) = \frac{1}{1+Exp(-x)}$

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
X = np.arange(-10, 10, .01)

F_of_X1 = lambda x: [np.exp(-0.5*((x-2)**2)) for x in X]
F_of_X2 = lambda x: [x**2 * np.exp(-0.8*((x+2)**2)) for x in X]
F_of_X3 = lambda x: [np.exp(-0.5*((x-2)**2)) * x**2 * np.exp(-0.8*((x+2)**2)) for x in X]
F_of_X4 = lambda x: [(1/(1+np.exp(-x))) for x in X]

fig, axes = plt.figure(figsize=(10,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.5)
ax1 = fig.add_subplot(2,2,1) #States there's 1 row of graphs, two columns of graphs, and we're selecting the 1st column
axes[0,1.plot(X, F_of_X1(X), color = 'red', label = 'F1')
ax1.legend(loc = 'best', fontsize = 'small')
ax1.set_xlabel('x')
ax1.set_ylabel('F(x)')
ax2 = fig.add_subplot(2,2,2) #Now selects the second graph (third number in the list)
ax2.plot(X, F_of_X2(X), color = 'red', label = 'F2')
ax2.legend(loc = 'best', fontsize = 'small')  #Python knows the 'best' spot in the graph to put the legend.
ax3 = fig.add_subplot(2,2,3) #Now selects the second graph (third number in the list)
ax3.plot(X, F_of_X3(X), color = 'red', label = 'F3')
ax3.legend(loc = 'best', fontsize = 'small')  #Python knows the 'best' spot in the graph to put the legend.
ax4 = fig.add_subplot(2,2,4) #Now selects the second graph (third number in the list)
ax4.plot(X, F_of_X4(X), color = 'red', label = 'F4')
ax4.legend(loc = 'best', fontsize = 'small')  #Python knows the 'best' spot in the graph to put the legend.
```

*** Exercise 2: ***
Plot the following functions and their respective contour plots in a 4x4 block. Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title.
1. $G_1(x,y) = Exp(-0.01*(x^3+y^3))$
2. $G_2(x,y) = Log(x^2+2*y^2)*x^2$


```python
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
Z = np.exp(-0.01*(X**3 + Y**3))
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1, projection = '3d')
ax1.plot_surface(X, Y, Z)
ax2 = fig.add_subplot(1,2,2)
ax2.contour(X, Y, Z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
Z = np.log(X**2 + 2 * Y**2)*X**2
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1, projection = '3d')
ax1.plot_surface(X, Y, Z)
ax2 = fig.add_subplot(1,2,2)
ax2.contour(X, Y, Z)
```

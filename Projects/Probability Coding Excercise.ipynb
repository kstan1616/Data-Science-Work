````python
from itertools import product
from matplotlib import pyplot as plt
%matplotlib inline

elements = [[0,1]] * 10
events = list(product(*elements))

successes = map(sum, events)
prob_success = lambda n: len([success for success in successes if success == n])
hist_list = [prob_success(x) for x in range(11)]

fig, ax = plt.subplots()
plt.plot(hist_list)
plt.show()

exp_value = lambda n: (len([success for success in successes if success == n]))/float(len(successes))


exp_list = [n*exp_value(n) for n in range(11)]
print sum(exp_list)


prob_zeroh = lambda n: (0.5)**(10-n)
print prob_zeroh(3)
prob_list = [prob_zeroh(t) for t in range(11)]
fig, ax = plt.subplots()
plt.plot(prob_list)
plt.show()





````

### 1)  Which employee had the highest total revenues?  Plot a bar chart to show this.

```python
df['Revenue'] = df['UnitPrice']*df['Quantity']
print df['Revenue']
pd.pivot_table(df, index = "EmployeeID", values = "Revenue", aggfunc= sum).plot(kind='bar')
```

### 2) Calculate and plot the total volume of products shipped by month

```python
import datetime
df['Month'] = df['ShippedDate'].apply(pd.to_datetime)
df["Month"] = df["Month"].apply(lambda x: x.month)
vol_month = pd.pivot_table(df, index = ['Month'], values = ['Quantity'], aggfunc = sum)
vol_month.plot(kind = 'bar')
```

### 3) Calculate and plot a histogram of the number of days taken to ship

```python
df['ShipDays'] = (df['ShippedDate'].apply(pd.to_datetime))-(df['OrderDate'].apply(pd.to_datetime))
df['ShipDays'].dt.days.hist()
```
